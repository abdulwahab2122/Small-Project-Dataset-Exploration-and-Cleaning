```python
train_dataset.fillna(train_dataset.mean(), inplace=True)
test_dataset.fillna(test_dataset.mean(), inplace=True)

train_dataset
test_dataset
```

```python
# Remove Duplicates from the dataset

train_dataset.drop_duplicates(inplace=True)
test_dataset.drop_duplicates(inplace=True)

train_dataset
test_dataset
```

```python
# Handle Outliers in the dataset

train_dataset.describe()

test_dataset.describe()
```

```python
# Remove Outliers from the dataset

train_dataset = train_dataset[train_dataset['Age'] < 80]
train_dataset = train_dataset[train_dataset['Fare'] < 300]

test_dataset = test_dataset[test_dataset['Age'] < 80]
test_dataset = test_dataset[test_dataset['Fare'] < 300]

train_dataset.describe()

test_dataset.describe()

# Data Visualization

sns.countplot(x='Survived', data=train_dataset)

```

```python
# Normalize or Standardize the numerical features

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

train_dataset[['Age', 'Fare']] = scaler.fit_transform(train_dataset[['Age', 'Fare']])
test_dataset[['Age', 'Fare']] = scaler.fit_transform(test_dataset[['Age', 'Fare']])

train_dataset

test_dataset
```

```python
# Divide the dataset into Dependent and Independent Variables

X_train = train_dataset.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'], axis=1)
y_train = train_dataset['Survived']

X_test = test_dataset.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

X_train

y_train

X_test

# Convert Categorical Features into Numerical Features

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

X_train

X_test

```

```python
# Encode Catetgorical Variables

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

X_train.iloc[:, 1] = encoder.fit_transform(X_train.iloc[:, 1])
X_train.iloc[:, 2] = encoder.fit_transform(X_train.iloc[:, 2])
X_train.iloc[:, 7] = encoder.fit_transform(X_train.iloc[:, 7])

X_train


```
